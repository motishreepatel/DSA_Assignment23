```python
#Q1 - 
'''  
Given preorder of a binary tree, calculate its **[depth(or height)](https://www.geeksforgeeks.org/write-a-c-program-to-find-the-maximum-depth-or-height-of-a-tree/)** [starting from depth 0]. The preorder is given as a string with two possible characters.

1. ‘l’ denotes the leaf
2. ‘n’ denotes internal node

The given tree can be seen as a full binary tree where every node has 0 or two children. The two children of a node can ‘n’ or ‘l’ or mix of both.

**Examples :**

Input  for this question: nlnll
Output for this question should be : 2
'''
#Answer:
def calculate_depth(preorder):
    depth = 0  # Initialize depth as 0
    stack = []  # Use a stack to keep track of nodes

    for char in preorder:
        if char == 'n':
            # Increment depth for internal nodes
            depth += 1
            stack.append(char)
        elif char == 'l':
            # Pop the corresponding internal nodes from the stack
            while stack and stack[-1] != 'n':
                stack.pop()

            # Pop the last internal node from the stack
            if stack and stack[-1] == 'n':
                stack.pop()

    return depth


# Given preorder traversal string
preorder = 'nlnll'

# Calculate the depth of the binary tree
depth = calculate_depth(preorder)

print("Depth of the binary tree:", depth)
```

```python
#Q2 - 
'''  
Given a Binary tree, the task is to print the **left view** of the Binary Tree. The left view of a Binary Tree is a set of leftmost nodes for every level.

**Examples:**

***Input for this question is:***

            4

          /   \

        5     2

             /   \

            3     1

           /  \

          6    7

Output for this question should be: 4 5 3 6

'''
#Answer:
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def print_left_view(root):
    if root is None:
        return

    max_level = [0]  # Store the maximum level reached so far
    print_left_view_util(root, 1, max_level)


def print_left_view_util(node, level, max_level):
    if node is None:
        return

    # If the current level is greater than the maximum level reached so far, print the node
    if level > max_level[0]:
        print(node.value, end=" ")
        max_level[0] = level

    # Recursively traverse the left and right subtrees
    print_left_view_util(node.left, level + 1, max_level)
    print_left_view_util(node.right, level + 1, max_level)


# Create the binary tree from the given input
root = Node(4)
root.left = Node(5)
root.right = Node(2)
root.right.left = Node(3)
root.right.right = Node(1)
root.right.left.left = Node(6)
root.right.left.right = Node(7)

# Print the left view of the binary tree
print("Left View of Binary Tree:")
print_left_view(root)
```

```python
#Q3 - 
'''   
Given a Binary Tree, print the Right view of it.

The right view of a Binary Tree is a set of nodes visible when the tree is visited from the Right side.

**Examples:**

**Input for this question is:**

         1

      /     \

   2         3

/   \       /  \

4     5   6    7

             \

               8

**Output for this question should be**: 

Right view of the tree is 1 3 7 8

'''

#Answer:
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def print_right_view(root):
    if root is None:
        return

    max_level = [0]  # Store the maximum level reached so far
    print_right_view_util(root, 1, max_level)


def print_right_view_util(node, level, max_level):
    if node is None:
        return

    # If the current level is greater than the maximum level reached so far, print the node
    if level > max_level[0]:
        print(node.value, end=" ")
        max_level[0] = level

    # Recursively traverse the right and left subtrees (right-to-left order)
    print_right_view_util(node.right, level + 1, max_level)
    print_right_view_util(node.left, level + 1, max_level)


# Create the binary tree from the given input
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)
root.right.right.right = Node(8)

# Print the right view of the binary tree
print("Right View of Binary Tree:")
print_right_view(root)
```

```python
#Q4 - 
'''  
Given a Binary Tree, The task is to print the **bottom view** from left to right. A node **x** is there in output if x is the bottommost node at its horizontal distance. The horizontal distance of the left child of a node x is equal to a horizontal distance of x minus 1, and that of a right child is the horizontal distance of x plus 1.

**Examples:**

**Input for this question is:**

             20

           /     \

        8         22

    /      \         \

5         3        25

        /    \

   10       14

**Output for this question should be:** 5, 10, 3, 14, 25.
'''

#Answer:
from collections import deque


class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


def print_bottom_view(root):
    if root is None:
        return

    # Dictionary to store the bottommost node at each horizontal distance
    bottom_view = {}

    # Queue to perform level order traversal
    queue = deque([(root, 0)])

    while queue:
        node, hd = queue.popleft()

        # Update the bottommost node at the current horizontal distance
        bottom_view[hd] = node.value

        # Enqueue the left and right children with their corresponding horizontal distances
        if node.left:
            queue.append((node.left, hd - 1))
        if node.right:
            queue.append((node.right, hd + 1))

    # Print the bottom view nodes in ascending order of their horizontal distances
    for hd in sorted(bottom_view.keys()):
        print(bottom_view[hd], end=" ")


# Create the binary tree from the given input
root = Node(20)
root.left = Node(8)
root.right = Node(22)
root.left.left = Node(5)
root.left.right = Node(3)
root.right.right = Node(25)
root.left.right.left = Node(10)
root.left.right.right = Node(14)

# Print the bottom view of the binary tree
print("Bottom View of Binary Tree:")
print_bottom_view(root)
```
